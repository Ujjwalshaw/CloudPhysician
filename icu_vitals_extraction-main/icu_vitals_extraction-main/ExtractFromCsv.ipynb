```python
!mkdir annotations
# Function to extract from csv
def extract_info_from_csv(path):
  thisdict =    {
  "HR": 0,
  "SBP": 1,
  "DBP": 2,
  "MAP":3,
  "SPO2":4,
  "RR":5,
  "HR_W":6,
  "RR_W":7,
  "SPO2_W":8,
  }
  df = pd.read_csv(path)
  # Initialise the info dict 

  # loop over the annotations and images, preprocess them and store in lists
  for ind in df.index:
      # print(df['image_name'][ind], df['HR'][ind])
      print_buffer = []
      for col in df.columns[1:]:
        if pd.isnull(df[col][ind]):
          print("Found a nan")
        else:
          i= [j for j in df[col][ind][1:-1].split(", ")]
          i= [float(k) for k in i]
          # Access bounding box coordinates
          annot = i
          print_buffer.append("{} {:.3f} {:.3f} {:.3f} {:.3f}".format(thisdict[col], annot[0], annot[1], annot[2], annot[3]))

        save_file_name = os.path.join("annotations", df["image_name"][ind].replace("jpeg", "txt"))

        # Save the annotation to disk
        #! touch "annotations/save_file_name.txt"
        print("\n".join(print_buffer), file=open(save_file_name, "w"))
          

print(extract_info_from_csv('/content/drive/MyDrive/Classification Dataset/BPL-EliteView-EV10-B_Meditec-England-A.csv'))
extract_info_from_csv('/content/drive/MyDrive/Classification Dataset/BPL-EliteView-EV100-C.csv')
extract_info_from_csv('/content/drive/MyDrive/Classification Dataset/BPL-Ultima-PrimeD-A.csv')
extract_info_from_csv('/content/drive/MyDrive/Classification Dataset/Nihon-Kohden-lifescope-A.csv')
```
